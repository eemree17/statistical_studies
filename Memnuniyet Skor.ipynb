```python
#kütüphaneler eklenmiştir.
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#veri seti yüklenmiştir.
os.chdir('C:/Users/emrem/OneDrive/Masaüstü/YTÜ Yüksek/İst.Bilg.Uyg/3.odev/')
data = pd.read_csv('toplamveri3.csv')
data.head()
```

```python
#okul değerlendirme kolonları seçilmiştir.
degerlendirme = data[['ANKETNO','K1','K29','K31','K32','K33','K34','K35','K36','K37','K38','K39','K40','K41','K42','K43','K44','K45','K67','K68','K69','K70','K71','K72','K73','K74','K75','K76','K77','K78','K79','K80','K81','K82','K109','K110']]
degerlendirme.head()
```

```python
#veri tipleri değiştirilmiştir.
for i in degerlendirme:
    degerlendirme[i] = pd.to_numeric(degerlendirme[i], errors='coerce') 
degerlendirme['ANKETNO']=data['ANKETNO'].apply(str)
degerlendirme['K1']=data['K1'].apply(str)

#Okul isimleri eklenmiştir.
degerlendirme['K1'].replace('1','Anadolu',inplace=True)
degerlendirme['K1'].replace('2','Ankara',inplace=True)
degerlendirme['K1'].replace('3','Başkent',inplace=True)
degerlendirme['K1'].replace('4','Çukurova',inplace=True)
degerlendirme['K1'].replace('5','Dokuz Eylül',inplace=True)
degerlendirme['K1'].replace('6','Ege',inplace=True)
degerlendirme['K1'].replace('7','Gazi',inplace=True)
degerlendirme['K1'].replace('8','Hacettepe',inplace=True)
degerlendirme['K1'].replace('9','İstanbul Ticaret',inplace=True)
degerlendirme['K1'].replace('10','Kadir Has',inplace=True)
degerlendirme['K1'].replace('11','Karadeniz Teknik',inplace=True)
degerlendirme['K1'].replace('12','Afyon Kocatepe',inplace=True)
degerlendirme['K1'].replace('13','Kırıkkale',inplace=True)
degerlendirme['K1'].replace('14','Mimar Sinan',inplace=True)
degerlendirme['K1'].replace('15','Muğla',inplace=True)
degerlendirme['K1'].replace('16','Ondokuz Mayıs',inplace=True)
degerlendirme['K1'].replace('17','Orta Doğu Teknik',inplace=True)
degerlendirme['K1'].replace('18','Osmangazi',inplace=True)
degerlendirme['K1'].replace('19','Selçuk',inplace=True)
degerlendirme['K1'].replace('20','Ufuk',inplace=True)
degerlendirme['K1'].replace('21','Yıldız Teknik',inplace=True)
degerlendirme['K1'].replace('22','Yaşar',inplace=True)
degerlendirme['K1'].replace('23','Fırat',inplace=True)

degerlendirme.head()
```

```python
#Boş veri sayıları ve oranları hesaplanmaştır.
total = degerlendirme.isnull().sum().sort_values(ascending = False)
percent = (degerlendirme.isnull().sum()/degerlendirme.isnull().count()*100).sort_values(ascending = False)
missing_data  = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing_data.head(20)
```

Boş veriler matematiksel hesaplamalarda sorunlara yol açmaktadır. Bu durumda 2 seçenek karşımıza çıkmaktadır. Veriler datadan silinmeli yada veri ile doldurulmalıdır. Gözlem sayısını azaltmamak adına verilerin doldurulmasına karar verilmiştir. Verilerin dağılıma etki etmemesi için ortalama değer atanmıştır.

```python
#Boş verilere ortalama değerler atanmıştır.
degerlendirme.fillna(degerlendirme.mean(), inplace=True)
degerlendirme.head()
```

```python
#Korelasyon Matrisi
f,ax = plt.subplots(figsize=(25, 25))
sns.heatmap(degerlendirme.corr(), annot=True, linewidths=.5, fmt= '.1f',ax=ax)
plt.show()
```

```python
#Sorular soru gruplarına göre ayrılmıştır.
#Anket ile verisetindeki soru kodları arasında farklılık tespit edilmiş, bu farklılık düzeltildikten sonra çalışmalara devam edilmiştir.
s8=degerlendirme[['K1','K29']]
s10=degerlendirme[['K1','K31','K32','K33','K34','K35','K36','K37','K38','K39']]
s11=degerlendirme[['K1','K40','K41','K42','K43','K44','K45']]
s23=degerlendirme[['K1','K67','K68','K69','K70','K71','K72','K73','K74','K75','K76','K77','K78','K79','K80','K81','K82']]
s28=degerlendirme[['K1','K109']]
s29=degerlendirme[['K1','K110']]
```

```python
#Soru gruplarına yapılan değerlendirmelerin ortalaması hesaplanmıştır.
s10["s10_mean"]=s10.loc[:,'K31':'K39'].mean(axis=1)
s11["s11_mean"]=s11.loc[:,'K40':'K45'].mean(axis=1)
s23["s23_mean"]=s23.loc[:,'K68':'K83'].mean(axis=1)
```

```python
#Okullara verilen ortalama puanlar hesaplanmıştır.
s8_mean=s8[['K1','K29']].groupby(['K1']).mean()
s10_mean=s10[['K1','s10_mean']].groupby(['K1']).mean()
s11_mean=s11[['K1','s11_mean']].groupby(['K1']).mean()
s23_mean=s23[['K1','s23_mean']].groupby(['K1']).mean()
s28_mean=s28[['K1','K109']].groupby(['K1']).mean()
s29_mean=s29[['K1','K110']].groupby(['K1']).mean()

#Tablolar birleştirilmiştir.
result = pd.concat([s8_mean, s10_mean, s11_mean, s23_mean, s28_mean, s29_mean], axis=1, join='inner')
result.head()
```

```python
#Soru kategorilerine göre korelasyon matrisi
f,ax = plt.subplots(figsize=(5, 5))
sns.heatmap(result.corr(), annot=True, linewidths=.5, fmt= '.1f',ax=ax)
plt.show()
```

Birçok değişkenden oluşan verisetine, tek değişken ile açıklanabilmesi adına PCA yapılmış ve veriseti 1 boyuta indirgenmiştir.

```python
#Soru kategorilerine göre PCA yapılmıştır.
PCA_means = result.iloc[:,0:5]
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
PCA_means = sc.fit_transform(PCA_means)  

from sklearn.decomposition import PCA

pca = PCA(n_components=1)  
PCA_means = pca.fit_transform(PCA_means)  
explained_variance = pca.explained_variance_ratio_ 

explained_variance
```

```python
print(pd.DataFrame(pca.components_,columns=result.iloc[:,0:5].columns,index = ['PC-1']))
```

```python
#PCA değeri ile okul bazından ortalama değerlendirmelerin ortalaması alınmış ve karşılaştırılmıştır.
result['PCA1'] = PCA_means[:, 0]
result["mean"]=result.loc[:,'K29':'K110'].mean(axis=1)
result_sorted = result.sort_values(by=['PCA1'])
result_sorted
```

Veriler PCA ile tek bir bileşene indirgenmiştir. Aşağıdaki grafik incelendiğinde oluşan PCA değişkeninin ortalama değeler ile paralel olduğu gözlemlenmiştir.

```python
#scatter plot (PCA değeri - ortalama değerlendirme puanı)
x = result['PCA1']
y = result["mean"]
plt.scatter(x, y)

z = np.polyfit(x, y, 1)
p = np.poly1d(z)
plt.plot(x,p(x),"r--")

plt.show()
```

```python
#PCA değeri 0-100 aralığına alınmış, skora son hali verilmiştir.
result['Score'] =(x-min(x))/(max(x)-min(x))*100
score_sorted = result.sort_values(by=['Score'],ascending = False)
score_sorted[["Score"]]
```
